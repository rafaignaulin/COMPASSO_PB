```python
#Exercicio 01

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

a = np.arange(-5,5)

b = sigmoid(a)

plt.plot(a, b, marker='o', color='#FEF510')
plt.show()
```

```python
#Exercicio 02

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

a = np.arange(-5,5)

b = sigmoid(a)
c = relu(a)


sigmoid = plt.plot(a, b, marker='o', color='#FEF510', label='Sigmoid')
relu = plt.plot(a, c, marker='x', color='#CDCDCD', label='Relu')

plt.legend()
plt.show()
```

```python
#Exercicio 06
import csv
# estrutura para guardar os dados
dataset = {'Actor': [],
'Total Gross': [],
'Number of Movies': [],
'Average per Movie': [],
'#1 Movie': [],
'Gross': []}

arq = open('actors.csv', "r")
reader = csv.reader(arq)
next(reader)

for i, row in enumerate(reader):

# parsing
    dataset['Actor'].append(row[0])
    dataset['Total Gross'].append(float(row[1]))
    dataset['Number of Movies'].append(int(row[2]))
    dataset['Average per Movie'].append(float(row[3]))
    dataset['#1 Movie'].append(row[4])
    dataset['Gross'].append(float(row[5]))

# convertendo tudo para np arrays
for k in dataset.keys():
    dataset[k] = np.array(dataset[k])

criterio = 'Number of Movies'
indice_top_5 = np.argsort(dataset[criterio])[::-1][:5]



plt.figure(figsize=(10, 6))
plt.bar(dataset['Actor'][indice_top_5], dataset['Number of Movies'][indice_top_5], color='royalblue')



plt.title('Relacao Artistas com mais Filmes')
plt.xlabel('Nome dos Artistas')
plt.ylabel('Quantidade de filmes')


plt.show()

```
