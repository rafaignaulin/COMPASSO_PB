```python
#Exercicio 02
df = pd.read_csv('../semana_4/googleplaystore.csv', header=0, )

df.drop(index=[10472], inplace=True)


#fifa.Value = fifa.Value.apply(lambda x : str(x).replace('€', '').replace('M', '000K').replace('K', '000').replace('.', ''))

    
df.Installs = df.Installs.apply(lambda x: np.int64(x.replace('+', '').replace(',', '')))
df.Reviews = df.Reviews.apply(lambda x: np.int64(x.replace('+', '').replace(',', '')))
df.Price = df.Price.apply(lambda x: np.float64(x.replace('$', '').replace(',', '')))

#1
top_5_installs = df.sort_values(by='Installs', ascending=False).head(6)
plt.figure(figsize=(15, 6))
plt.bar(top_5_installs['App'].values, top_5_installs['Installs'].values, color='royalblue')
plt.title("Os top 5 aplicativos com maior quantidade de instalaçoes")



#2
plt.figure(figsize=(15, 15))
df.Category.value_counts().plot.pie(title="Todas as categorias", autopct='%1.1f%%')


#3
value = df.Price.idxmax()
print("3- O aplicativo mais caro do dataset é o", df.App[value], "custando", df.Price[value], "reais")

#4
cr = df['Content Rating'].value_counts()
print("4- Existem", cr['Mature 17+'], "linhas com Mature 17+.")

#5
top_10_review = df.sort_values(by='Reviews', ascending=False).head(10)
print_dataframe(top_10_review[['App', 'Reviews']])
```
