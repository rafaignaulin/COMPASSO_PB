```python
#Exercicio 03

dataset = np.loadtxt('../semana_4/wine_data.csv', delimiter=',')
dataset = dataset[:, :4] # ['Class label', 'Alcohol', 'Malic acid']
print(dataset.shape)

media_alcool = np.mean(dataset[:, 1])
media_acido = np.mean(dataset[:, 2])
media_vine = np.mean(dataset[:, 3])

#print(dataset[:, 1], dataset[:, 2]) #Antes de normalizar
dataset[:, 1] = dataset[:, 1] - media_alcool
dataset[:, 2] = dataset[:, 2] - media_acido
dataset[:, 3] = dataset[:, 3] - media_vine

#print(dataset, dataset[:, 1], dataset[:, 2]) #Normalizou

#dataset[:,] = np.linalg.norm(dataset[:,], axis=1, keepdims=True)

plt.scatter(dataset[:, 1], dataset[:, 2], c=dataset[:, 0])
plt.title('Alcohol and Malic Acid content of the wine dataset')
plt.xlabel('Alcohol')
plt.ylabel('Malic acid')
plt.show()
```

```python
#Exercicio 04
def plot_triangle(vertices, center=None):
    trishape = plt.Polygon(vertices, edgecolor='r', alpha=0.2, lw=5)
    _, ax = plt.subplots(figsize=(4, 4))
    ax.add_patch(trishape)
    ax.set_ylim([.5, 3.5])
    ax.set_xlim([.5, 3.5])
    if center is not None:
        ax.scatter(*center, color='g', marker='D', s=70)
    ax.scatter(vertices[:, 0], vertices[:, 1], color='b', s=70)
    plt.show()
    
    
coordenadas = np.array([[1, 1],
                        [3, 1],
                        [2, 3]])


x = np.mean(coordenadas[:,0])
y = np.mean(coordenadas[:,0])
print(x, y)
plot_triangle(coordenadas, center=(x, y))
```
