```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.sql import functions as f
from pyspark.sql import types as t
```

```python
spark = SparkSession.builder.appName("Twitter Transformation").getOrCreate()
tweets = t.StructType([
    t.StructField("id", t.StringType(), False),
    t.StructField("text", t.StringType(), False),
    t.StructField("created_at", t.StringType(), False)
])
df = spark \
    .read \
    .schema(tweets) \
    .json('/home/rafael_ignaulin/Desktop/COMPASSO/Sprint_5/week_10/tweet_streaming_raw/twitter_20210725_221253.json')
#.option("includeTimestamp", "True")\  
```

```python
df.show(100, False)
```

```python
dest = '/home/rafael_ignaulin/Desktop/COMPASSO/Sprint_5/week_10/tweet_streaming_refined'
query = df \
    .writeStream \
        .outputMode("append").format("csv")\
        .option("path", dest)\
        .option("checkpointLocation", dest + "/check")\
    .start()

query.awaitTermination()
```
